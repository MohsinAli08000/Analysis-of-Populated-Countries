```python
popcntr=popcntr.drop(columns=['Unnamed: 67', 'Country Name', 'Indicator Code', 'Indicator Name'])
popcntr
```

```python
for i in popcntr['Country Code']:
    null_count=(popcntr[popcntr['Country Code']==i]).isna().sum().sum()
    if null_count>0: # Filtering out the country code with na values
        print(i,null_count)
```

```python
row_INX=popcntr[popcntr['Country Code']=='INX']
row_PSE=popcntr[popcntr['Country Code']=='PSE']
```

```python
#filtering out rows with specific country code form population DataFrame 
#This DataFrame contains populations data excluding the country code INX and PSe
filtered_population=popcntr[popcntr['Country Code'] != 'INX']
filtered_population=filtered_population[filtered_population['Country Code'] !='PSE']
```

```python
filtered_population.isna().sum()
```

```python
filtered_population
```

```python
#find the value of minimum value in the DataFrame exluding the first column (country Code)
min_index=filtered_population.iloc[:,1:].stack().idxmin()

#extract the corresponding country code and year from the index
lowest_population_country,lowest_population_year=min_index

# retrieve the lowest population using the index
lowest_population=filtered_population.loc[lowest_population_country,lowest_population_year]

print("Lowest Populated Country:",lowest_population_country)
print("Total Population:",lowest_population)
print("Year:",lowest_population_year)
```

```python
# Finding the miximum value 
max_index=filtered_population.iloc[:,1:].stack().idxmax()

highest_population_country,highest_population_year=max_index

highest_population=filtered_population.loc[highest_population_country,highest_population_year]

print("Highest Populated Country:",highest_population_country)
print("Total Population:",highest_population)
print("Year:",highest_population_year)
```

```python
highest_population_country=filtered_population['Country Code'][highest_population_country]
lowest_population_country=filtered_population['Country Code'][lowest_population_country]
```

```python
# Removing country code from filtered population 

df_nocountry_code=filtered_population.drop(columns=['Country Code'])
```

```python
average_population=df_nocountry_code.mean(axis=1)
```

```python
#concate the country code from averagae population 
population_result = pd.concat([filtered_population['Country Code'], average_population], axis=1)
population_result.columns=['Country Code','Average Population'] 
```

```python
# for remove floating point from decimal number
pd.set_option('display.float_format', lambda x:'%.f' %x)

highest_population=population_result.sort_values(by='Average Population', ascending =False)
print(highest_population[:10])
```

```python
highest_population_country=highest_population['Country Code'][:10]
highest_population=highest_population['Average Population'][:10]
```

```python
# Plotting a pie chart to visualize the population distribution among the highest populated countries
plt.figure(figsize=(8, 8))
plt.pie(highest_population, labels=highest_population_country, autopct='%1.1f%%', colors=plt.cm.Paired.colors, startangle=140)
plt.title('Population Distribution: Highest Populated Countries')
plt.tight_layout()
plt.show()
```

This pie chart provides a visual representation of the population distribution among the highest populated countries, each identified by its unique country code. Here's what you need to know:

The pie slices represent the population sizes of different countries or regions. The size of each slice corresponds to the proportion of the total population that belongs to each country.

Labels: Each slice is labeled with a country code, such as EAS, LTE, LMC, UMC, EAR, IBD, MIC, LMY, IBT, and WLD, representing different countries or regions.

Percentages: The percentage on each slice indicates the share of the global population for that specific country or region.

This pie chart provides an easy-to-read comparison of how the population is distributed among the world's highest populated countries or regions.population Distribution Pie Chart - Highest Populated Countries


```python
plt.figure()

# Generating a color palette with the same number of colors as there are countries
colors = plt.cm.viridis(np.linspace(0, 1, len(highest_population_country)))

# Plotting a horizontal bar chart with population on the x-axis and country code on the y-axis
plt.barh(highest_population_country, highest_population, color=colors)
plt.xlabel('Population')
plt.ylabel('Country Code')
plt.title('Population Distribution: Highest Populated Countries')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Sort the DataFrame by 'Average Population' column in ascending order to find the countries with the lowest population
# Print the top 10 countries with the lowest average population
lowest_population = population_result.sort_values(by='Average Population', ascending=True)
print(lowest_population[:10])
```

```python
# Extract the top 10 countries with the lowest population from the DataFrame
lowest_population_country = lowest_population['Country Code'][:10]
lowest_population = lowest_population['Average Population'][:10]
```
